```python
import numpy as np

def check_win(board, user_value):
    for i in range(0, 3):
        if np.all(board[i] == user_value) or np.all(board[:, i] == user_value):
            return True
    if np.all(np.diag(board) == user_value) or np.all(np.diag(np.fliplr(board)) == user_value):
        return True
    return False

def check_draw(board):
    return " " not in board

def game_starts():
    board = np.full((3, 3), " ")
    player1 = input("Enter your name player 1: ")
    player2 = input("Enter your name player 2: ")
    check = player1
    while True:
        if check == player1:
            print(player1, "It's your choice")
        else:
            print(player2, "It's your choice")
        print(board)
        row = int(input("Enter row: "))
        col = int(input("Enter column: "))
        if row not in range(0, 3) or col not in range(0, 3):
            print("Invalid input. Row and column should be in range 0 to 2.")
            continue
        if board[row][col] != " ":
            print("It is already occupied")
            continue

        if check == player1:
            board[row][col] = 'x'
            if check_win(board, 'x'):
                print(board)
                print(player1, "won")
                break
        else:
            board[row][col] = 'o'
            if check_win(board, 'o'):
                print(board)
                print(player2, "won")
                break

        if check_draw(board):
            print(board)
            print("It's a draw!")
            break

        if check == player1:
            check = player2
        else:
            check = player1

print("Tic Tac Toe")
play1 = input("Are you ready? (yes or no): ")
if play1 == "yes":
    game_starts()
else:
    print("You chose not to play")
```
